```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("training.csv")
```

```python
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
df_transformed = data.groupby("prognosis").max().reset_index()
```

```python
df_transformed
```

```python
df_transformed['prognosis'].unique()
```

```python
df_transformed.nunique()
```

```python
encoder = LabelEncoder()
df_transformed['prognosis'] = encoder.fit_transform(df_transformed['prognosis']) 
```

```python
df_transformed
```

```python
label_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_))) 
```

```python
label_mapping
```

```python
inverse_mapping = {v: k for k, v in label_mapping.items()}
```

```python
inverse_mapping
```

```python
testing_df = pd.read_csv("testing.csv")
```

```python
testing_df
```

```python
testing_df.replace({'prognosis': label_mapping}, inplace=True)
```

```python
testing_df
```

```python
X_train = df_transformed.drop(columns=['prognosis'])
y_train = df_transformed['prognosis']
X_test = testing_df.drop(columns=['prognosis'])
y_test = testing_df['prognosis']
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_rf)
```

```python
svm = SVC(kernel='linear', probability=True)
```

```python
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_svm)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.model_selection import cross_val_score
```

```python
accuracy = accuracy_score(y_test, y_pred_rf)
precision = precision_score(y_test, y_pred_rf, average="weighted")
recall = recall_score(y_test, y_pred_rf, average="weighted")
f1 = f1_score(y_test, y_pred_rf, average="weighted")
```

```python
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

```python

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

accuracy_score(y_test, y_pred)
```
